Determine the job_title based on years of job experience and years of SQL experience. The variable job_experience is set to 4 and sql_experience is set to 2. Use the following criteria:

Less than 2 years of job experience:

1. 0 SQL experience → "Associate Analyst"
2. 1+ years of SQL experience → "Junior Analyst"

2 to 5 years of job experience:

1. 0 SQL experience → "Associate Analyst"
2. 1 year of SQL experience → "Intermediate Analyst"
3. 2+ years of SQL experience → "Senior Analyst"

More than 5 years of job experience:

1. 0–2 years of SQL experience → "Data Analyst"
2. 3+ years of SQL experience → "Lead Analyst"

```python
job_experience = 4
sql_experience = 2

if job_experience <2:
  if sql_experience ==0: # in conditional statement, it's == (equal to ), not = (assgin)
    job_title = "Associate Analyst"
  else:
    job_title = "Junior Analyst"
elif job_experience >=2 and job_experience <=5:
  if sql_experience ==0:
    job_title = "Associate Analyst"
  elif sql_experience ==1:
    job_title = "Intermediate Analyst"
  else:
    job_title = "Senior Analyst"
elif job_experience > 5: #here cannot be else
  if sql_experience <=2:
    job_title = "Data Analyst"
  elif sql_experience > 3:
    job_title = "Lead Analyst"
print(job_title)


```

# ***Cheat Sheet***

List Method

| Method              | Behavior         | Returns          | Example                     |
| ------------------- | ---------------- | ---------------- | --------------------------- |
| `.append(x)`        | In-place         | `None`           | `lst.append(10)`            |
| `.extend(iterable)` | In-place         | `None`           | `lst.extend([1, 2])`        |
| `.insert(i, x)`     | In-place         | `None`           | `lst.insert(1, 'a')`        |
| `.remove(x)`        | In-place         | `None`           | `lst.remove('a')`           |
| `.pop([i])`         | In-place         | Removed element  | `x = lst.pop()`             |
| `.clear()`          | In-place         | `None`           | `lst.clear()`               |
| `.sort()`           | In-place         | `None`           | `lst.sort()`                |
| `sorted(lst)`       | Returns new list | New sorted list  | `sorted_list = sorted(lst)` |
| `.reverse()`        | In-place         | `None`           | `lst.reverse()`             |
| `reversed(lst)`     | Returns iterator | New reverse view | `for x in reversed(lst):`   |

| Method               | Behavior    | Returns         | Example                    |

| `.remove(x)`         | In-place    | `None`          | `s.remove('a')`            |
| `.discard(x)`        | In-place    | `None`          | `s.discard('a')`           |

| `.union(set)`        | Returns new | Combined set    | `s3 = s1.union(s2)`        |
| `.intersection(set)` | Returns new | Shared elements | `s3 = s1.intersection(s2)` |
| `.difference(set)`   | Returns new | Difference set  | `s3 = s1.difference(s2)`   |

**DataFrame.drop**(labels=None, *, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')

Remove rows or columns by specifying label names and corresponding axis, or by directly specifying index or column names. When using a multi-index, labels on different levels can be removed by specifying the level.

1. labels
single label or list-like
Index or column labels to drop. A tuple will be used as a single label and not treated as a list-like.
2. axis
{0 or ‘index’, 1 or ‘columns’}, **default 0**
Whether to drop labels from the index (0 or ‘index’) or columns (1 or ‘columns’).
3. index
single label or list-like
Alternative to specifying axis (labels, axis=0 is equivalent to index=labels).
4. columns
single label or list-like
Alternative to specifying axis (labels, axis=1 is equivalent to columns=labels).
5. level
int or level name, optional
For MultiIndex, level from which the labels will be removed.
6. inplace
bool, **default False**
If False, return a copy. Otherwise, do operation in place and return None.
7. errors
{‘ignore’, ‘raise’}, **default ‘raise’**
If ‘ignore’, suppress error and only existing labels are dropped.

**DataFrame.dropna**(*, axis=0, how=<no_default>, thresh=<no_default>, subset=None, inplace=False, ignore_index=False)

1. axis{0 or ‘index’, 1 or ‘columns’}, **default 0**
Determine if rows or columns which contain missing values are removed.
0, or ‘index’ : Drop rows which contain missing values.

Only a single axis is allowed.

Determine if row or column is removed from DataFrame, when we have at least one NA or all NA.
‘any’ : If any NA values are present, drop that row or column.
‘all’ : If all values are NA, drop that row or column.
threshint, optional
Require that many non-NA values. Cannot be combined with how.
3. subsetcolumn label or sequence of labels, optional
Labels along other axis to consider, e.g. if you are dropping rows these would be a list of columns to include.
4. inplacebool,** default False**
Whether to modify the DataFrame rather than creating a new one.
5. ignore_indexbool, **default False**
If True, the resulting axis will be labeled 0, 1, …, n - 1.
