```python
df.loc[0:100,'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate')
```

- df.loc[0:100, 'salary_rate':'salary_hour_avg'] returns a **new DataFrame slice** (a copy or view of that part of the original DataFrame).
- .dropna(subset=['salary_rate']) returns a new DataFrame where rows with missing 'salary_rate' in that slice are dropped.
- Since inplace=False by default, it does not modify the original DataFrame, just returns the cleaned slice.
- So, the whole expression evaluates to a new smaller DataFrame (the cleaned slice).

```python
df.loc[0:100,'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate')
```

Why does it return nothing?  
Methods with **inplace=True** always return **None** to signal that they modify the object directly instead of creating and returning a new one.

So even though the slice is modified, the expression itself doesn't return anything you can assign or print.

```python
df.loc[0:100,'salary_rate':'salary_hour_avg'].dropna(subset='salary_rate', inplace=True)
df.loc[0:100,'salary_rate':'salary_hour_avg']
```

| Use Case                       | `inplace=True`   | `inplace=False`                    |
| ------------------------------ | ---------------- | ---------------------------------- |
| Want a modified copy           | ‚ùå Returns `None` | ‚úÖ Returns the modified DataFrame   |
| Want to update existing object | ‚úÖ Works directly | ‚úÖ Assign back to the same variable |

# When to use Inplace=True

## ‚úîÔ∏è Example 1: Dropping rows in place

```python
df.dropna(subset=['salary_rate'], inplace=True)
"""
This will remove rows from df where 'salary_rate' is NaN.
df is now changed.
Nothing is returned (i.e., the method returns None).
"""
```

```python
df.head()
```

## ‚úîÔ∏è Example 2: Sorting in place

df.sort_values(by='salary_rate', inplace=True)

## ‚úîÔ∏è Example 3: Renaming columns in place

üß† Best Practice (Recommended)

Even though inplace=True works, Pandas developers recommend not using it in most cases.

Why?
- It's less clear.
- May lead to unexpected side effects.
- It's being phased out in some cases.

- df.loc[0, 'salary_rate']         # Row with index label 0, column 'salary_rate'
- df.loc[0:3, 'salary_rate':'salary_hour_avg']  # Slicing by label
- df.loc[df['salary_rate'] > 100]  # Conditional selection


```python
df_subset= df.dropna(subset='salary_year_avg')
df_subset.loc[df_subset['salary_year_avg']>100000]
```

```python
df.loc[df['salary_year_avg']>100000]
```

| Expression                            | Behavior                       |
| ------------------------------------- | ------------------------------ |
| `df['col'] > value`                   | Returns `True`/`False`/`NaN`   |
| `df.loc[df['col'] > value]`           | Drops `NaN` rows automatically |
| Want to drop `NaN` explicitly before? | Use `.dropna(subset=['col'])`  |


## ‚úÖ Summary: Methods that modify in place

| Object Type | Common In-place Methods                                                                                           |
| ----------- | ----------------------------------------------------------------------------------------------------------------- |

| Dict        | clear, pop, popitem, update, setdefault                                                                           |

## .loc[] used as a filter method

1. **Filtering by boolean condition**:  

df.loc[**df['salary_year_avg']** > 100000]

2. **Filtering by index labels**:  

df.loc[['United States', 'Canada']]

3. **.loc[] can do row and column selection together** :

df.loc[df['job_country'] == 'United States', ['job_title', 'salary_year_avg']]
- This filters rows where job_country is 'United States' and selects just the specified columns.

## ‚úÖ Example

```python
# You're not applying a filter ‚Äî you're selecting rows by index label.
df.loc[['United States', 'Canada']]
```

- This only works if 'United States' and 'Canada' are actual index values in the DataFrame.
- You are saying: ‚ÄúGive me the rows whose index is exactly 'United States' or 'Canada'.‚Äù

```python
import pandas as pd

df = pd.DataFrame({
    'job_country': ['United States', 'Canada', 'Germany', 'United States'],
    'salary': [100000, 90000, 80000, 120000]
})

# Set 'job_country' as the index
df = df.set_index('job_country')

# Now this works:
df.loc[['United States', 'Canada']]
```

## ‚ùå If the index is not set:
If 'job_country' is just a column (not the index), this won‚Äôt work:

```python
# initially, the DataFrame you create doesn't have a custom index. 
# It just uses the default integer index: 0, 1, 2, ....
import pandas as pd

df = pd.DataFrame({
    'job_country': ['United States', 'Canada', 'Germany', 'United States'],
    'salary': [100000, 90000, 80000, 120000]
})


# This will raise a KeyError
df.loc[['United States', 'Canada']]
```
