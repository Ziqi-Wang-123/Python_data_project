## what random_state does


What to expect:
- Without random_state: The sampled rows will change each time you run the code.
- With random_state=42: The sampled rows will be **exactly the same every run**.

| Parameter      | Type                                                          | Default | Description                                                                 | Example                                                  |
| -------------- | ------------------------------------------------------------- | ------- | --------------------------------------------------------------------------- | -------------------------------------------------------- |
| `n`            | int or None                                                   | None    | Number of items to return. Cannot be used with `frac`.                      | `df.sample(n=5)`  — sample 5 rows                        |
| `frac`         | float or None                                                 | None    | Fraction of axis items to return. Cannot be used with `n`.                  | `df.sample(frac=0.2)` — sample 20% of rows               |
| `replace`      | bool                                                          | False   | Sample with or without replacement.                                         | `df.sample(n=5, replace=True)` — sample with replacement |
| `weights`      | str, ndarray, Series, or None                                 | None    | Weights for sampling rows. Must be non-negative and sum to 1 (if provided). | `df.sample(n=5, weights='prob_col')` — weighted sample   |
| `random_state` | int, array-like, BitGenerator, np.random.RandomState, or None | None    | Seed or random number generator for reproducibility.                        | `df.sample(n=5, random_state=42)` — reproducible sample  |
| `axis`         | {0 or ‘index’, 1 or ‘columns’}                                | 0       | Axis to sample — 0 or ‘index’ for rows, 1 or ‘columns’ for columns.         | `df.sample(n=2, axis=1)` — sample 2 columns              |
| `ignore_index` | bool                                                          | False   | If True, the resulting axis will be labeled 0, 1, …, n - 1.                 | `df.sample(n=5, ignore_index=True)` — reset index        |


## What is replacement in sampling?  
Replacement refers to whether you allow the same item to be picked more than once when sampling.

- Without replacement (replace=False):  
Once an item (row) is picked, it cannot be picked again.  
**So all sampled items are unique.**
- With replacement (replace=True):  
After picking an item, it is put back, so it can be picked again.  
This means some rows may appear **multiple times** in the sample.

## What ignore_index does:  
When you sample rows (or do operations that return a subset), the original row indices are preserved by default.
- If ignore_index=True, pandas resets the index in the result to a default integer sequence: 0, 1, 2, ....
- If ignore_index=False, the sampled DataFrame **keeps the original row labels/indices**.
