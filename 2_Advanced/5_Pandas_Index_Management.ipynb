### reset_index()

ðŸ“˜ reset_index() Parameters Overview

| Parameter        | Type              | Default | Description                                                      | Example               |
| ---------------- | ----------------- | ------- | ---------------------------------------------------------------- | --------------------- |
| `level`          | int, str, or list | `None`  | Only remove specific level(s) from a MultiIndex                  | `level='region'`      |
| `drop`           | bool              | `False` | If `True`, removes the index and doesnâ€™t add it back as a column | `drop=True`           |
| `inplace`        | bool              | `False` | If `True`, modifies the DataFrame in place                       | `inplace=True`        |
| `names` (>=v2.1) | list              | `None`  | New names for the reset index (only in pandas 2.1.0+)            | `names=['new_index']` |


df.reset_index(names=['new_name1', 'new_name2', ...])
- names is a list of strings corresponding to new column names replacing the original index level names.

#### The name argument is for the reset_index() method of a pandas Series, NOT the DataFrame one.

####  For Series, not dataframe
reset_index(name='')

- reset_index() alone turns Series â†’ DataFrame with default numeric column name for values.
- reset_index(name=...) turns Series â†’ DataFrame and assigns a meaningful column name to values.

**reset_index() converts a Series into a DataFrame by turning the current index into a regular column and creating a new default integer index (0, 1, 2, ...).**

| Method                                | Has `name` param? | What it does                                         |
| ------------------------------------- | ----------------- | ---------------------------------------------------- |
| `pandas.Series.reset_index(name=...)` | Yes               | Converts Series to DataFrame and names values column |
| `pandas.DataFrame.reset_index()`      | No                | Resets index, turning index into columns             |


```python
df_us=df[df['job_country']=='United States']
df_us.reset_index(inplace=True)
df_us
```

### set_index()

ðŸ“˜ set_index() Parameters Overview

| Parameter          | Type               | Default | Description                                                                | Example                         |
| ------------------ | ------------------ | ------- | -------------------------------------------------------------------------- | ------------------------------- |
| `keys`             | str or list of str | â€”       | Column(s) to set as the new index                                          | `'job_country'` or `['A', 'B']` |
| `drop`             | bool               | `True`  | Whether to drop the column(s) used as index from the DataFrame             | `drop=False` keeps the column   |
| `append`           | bool               | `False` | Whether to append columns to existing index (create a MultiIndex)          | `append=True`                   |
| `inplace`          | bool               | `False` | Whether to modify the DataFrame in place                                   | `inplace=True`                  |
| `verify_integrity` | bool               | `False` | Checks for duplicate entries in the new index. Errors if duplicates found. | `verify_integrity=True`         |

Create a pivot table to calculate the median salary_year_avg for each job_title_short, reset the index of the resulting pivot table, then set the index to salary_year_avg. Display the first 5 rows.

```python
df_pivot1=df.pivot_table(values='salary_year_avg',index='job_title_short',aggfunc='median')

df_pivot1.reset_index(inplace=True)
df_pivot1.set_index('salary_year_avg',inplace=True)
df_pivot1.head()
```

## âœ… So why reset then set?

1. Initially, 'job_title_short' was the index.
2. You can't directly set 'salary_year_avg' as the new index unless it's a column.
3. So, you use .reset_index() to turn 'job_title_short' into a column, making 'salary_year_avg' available as a column too.
4. Then you can safely use .set_index('salary_year_avg').

```python
df_pivot2=df.pivot_table(values='salary_year_avg',index='job_title_short',aggfunc='median')

df_pivot2.columns
```
