2. Re-analyze based on percentage of total jobs
3. Plot the monthly skill demand

## Import Libraries and Data

```python
# import libraries
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns

#import data
df=pd.read_csv('/Users/abby/Desktop/Python_Data_Analytics_Luke/data_jobs.csv')

# data cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
# Convert string representation to actual list, checking for NaN values first
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

 ## Group the Skills by Month

Select only those job postings that are for Data Analysts and the job country is the United States. Then extract the month from each job's posting date to see when jobs are listed. Next, converts a column that lists skills into a usable list format. Finally, it rearranges the data so that each skill from the list gets its own row.

```python
## Skill count per month for data analysts
df_DA_US = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()

df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month

# Explode job_skills column
df_DA_US_explode = df_DA_US.explode('job_skills')
```

### Pivot in Prep for Plotting

Create a pivot table from the `df_da_us_explode`, setting 'month' as the index, `job_skills` as the columns, and fills missing entries with zero.

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no',columns='job_skills',aggfunc='size',fill_value=0)
df_DA_US_pivot 
```

#### Sort columns by count and change month numbers to names

It adds a new row labeled `Total` that sums up counts across all months for each skill. Finally, it reorders the columns based on the total counts, displaying them from highest to lowest, and shows the updated pivot table. 

```python
df_DA_US_pivot.loc['Total']=df_DA_US_pivot.sum()
```

```python
df_DA_US_pivot['airflow']
```

```python
# sorts by count

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]


df_DA_US_pivot
```

 ## Plot the Monthly Skill Counts

Plot a line chart of the top 5 skills for data analysts, indexed by month. It selects the first five columns and plots them. 

```python

# Drop Totals before ploting
df_DA_US_pivot = df_DA_US_pivot.drop('Total')

df_DA_US_pivot.iloc[:, :5].plot(kind='line')

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

This is good but totals can be misleading without context. Percentages clarify relationships within the whole, giving a more accurate comparison.

## Calculate Percentage of Total Jobs

```python
# Get monthly totals
DA_totals = df_DA_US.groupby('job_posted_month_no').size()

DA_totals
```

Then calculate the percentage of each skill's occurrence for each month by dividing the first 12 rows of the `df_da_us_pivot` DataFrame, which contains skill counts, by the monthly totals in `DA_totals` and multiplying by 100. The new DataFrame, `df_DA_pivot_percent`, shows skill distributions as percentages of total monthly job entries.

```python
# divide first 12 rows of df_DA_US_pivot by DA_totals
df_DA_US_percent = df_DA_US_pivot.iloc[:12].div(DA_totals/100, axis=0)
df_DA_US_percent
```

## 🧠 What's happening:  

- `df_da_us_pivot.iloc[:12]`   Selects the first 12 rows of the DataFrame df_da_us_pivot.   
- `.div(DA_totals / 100, axis=0)` Divides each row in the selected data by the corresponding value in DA_totals / 100, row-wise (axis=0 = match by index).

```python

# changes month number to month name

# Step 1: Reset the index
df_DA_US_percent = df_DA_US_percent.reset_index()

# Step 2: Convert numeric month to abbreviated name
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))

# Step 3: Set new month names as the index
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')

# Step 4: Drop the old numeric month column
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')

df_DA_US_percent
```

## Plot Monthly Skill Demand 

Creates a line plot for the top five skills of data analysts, shown as percentages of the total job entries per month, using the first 5 columns of the `df_da_pivot_percent` DataFrame. Also the legend is moved outside of the plot for readability.

```python
# select first five columns to plot
df_plot = df_DA_US_percent.iloc[:,:5]

sns.lineplot(data=df_plot,dashes=False, legend='full',palette='tab10')
sns.set_theme(style='ticks')
sns.despine() #remove top and right spines

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y,pos:f'{y:.0f}%'))

# annotate the plot with the top 5 skills using plt.text()
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i], color='black')


# 11.2: This is the x-coordinate where the text will appear. It’s likely outside or to the right of the plotted data range (e.g., 12 months), so the label sits just outside the final point on the x-axis.
# df_plot.iloc[-1, i]:
# This is the y-coordinate—the value of the last row (-1) in column i. So you're placing the label aligned with the latest data point for each line.
# df_plot.columns[i]:
# This is the actual text label, probably the skill name (e.g., 'SQL', 'Python', etc.).

plt.show()
```

## Another way to manage the format of Y-axis

```python
from matplotlib.ticker import PercentFormatter

df_plot = df_DA_US_percent.iloc[:, :5]
sns.lineplot(data=df_plot, dashes=False, legend='full', palette='tab10')
sns.set_theme(style='ticks')
sns.despine() # remove top and right spines

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

# annotate the plot with the top 5 skills using plt.text()
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i], color='black')

plt.show()
```
